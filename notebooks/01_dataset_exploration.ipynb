# 01 — Dataset Exploration

This notebook loads the raw CICIDS2017 Wednesday dataset and performs initial exploratory data analysis:
- Inspect shape, types, and head of the data
- Examine class distribution (normal vs DoS attacks)
- Compute feature-level statistics
- Detect missing and infinite values
- Visualize feature correlations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

sns.set_style("whitegrid")
plt.rcParams["figure.dpi"] = 100
```

## 1. Load Raw Dataset

```python
sys.path.insert(0, os.path.join(os.path.dirname("__file__" if "__file__" in dir() else ""), ".."))

RAW_PATH = os.path.join("..", "data", "raw", "CICIDS2017", "Wednesday-workingHours.pcap_ISCX.csv")
df = pd.read_csv(RAW_PATH)

print(f"Shape: {df.shape}")
print(f"Columns: {df.shape[1]}")
df.head()
```

```python
df.dtypes
```

## 2. Class Distribution

```python
print(df["Label"].value_counts())
df["Label"].value_counts().plot(kind="barh", figsize=(10, 6))
plt.xlabel("Count")
plt.ylabel("Label")
plt.title("Class Distribution — Wednesday CICIDS2017")
plt.tight_layout()
plt.show()
```

## 3. Feature Statistics

```python
df.describe()
```

```python
df.info()
```

## 4. Missing Values

```python
null_counts = df.isnull().sum()
print("Null values per column:")
print(null_counts[null_counts > 0])
print(f"\nTotal null values: {null_counts.sum()}")

inf_counts = df.isin([np.inf, -np.inf]).sum()
print("\nInfinite values per column:")
print(inf_counts[inf_counts > 0])
print(f"\nTotal infinite values: {inf_counts.sum()}")
```

## 5. Feature Correlations

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(corr, cmap="coolwarm", center=0, ax=ax, xticklabels=False, yticklabels=False)
ax.set_title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()
```

## 6. Summary

```python
total_nulls = df.isnull().sum().sum()
total_infs = df.isin([np.inf, -np.inf]).sum().sum()
n_labels = df["Label"].nunique()

print("=" * 50)
print("DATASET EXPLORATION SUMMARY")
print("=" * 50)
print(f"Rows:                {df.shape[0]:,}")
print(f"Columns:             {df.shape[1]}")
print(f"Numeric features:    {len(numeric_cols)}")
print(f"Unique labels:       {n_labels}")
print(f"Total null values:   {total_nulls}")
print(f"Total infinite vals: {total_infs}")
print("\nClass distribution:")
for label, count in df["Label"].value_counts().items():
    pct = count / len(df) * 100
    print(f"  {label:<25s} {count:>10,}  ({pct:.2f}%)")
print("=" * 50)
```
